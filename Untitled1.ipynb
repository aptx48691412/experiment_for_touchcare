```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
```

```python
df=pd.read_csv('./csv/speed__1220.csv',header=None)
df
```

```python
name_list=list(df.iloc[:,0].unique())[:-1]
name_list=name_list[:4]+name_list[5:]
name_list
```

```python
df_=df[df.iloc[:,0].isin(name_list)]
stroke,speak=df_.iloc[:,1],df_.iloc[:,2]
```

```python
stroke_list=list()
speak_list=list()

for i,k in zip(stroke,speak):
    if int(i)!=0 and int(k)!=0 and k<10:
        stroke_list.append(i)
        speak_list.append(k)
```

```python
from sklearn.linear_model import LinearRegression


fig=plt.figure(dpi=150)

model=LinearRegression()
model.fit(np.array(stroke_list).reshape(-1,1),np.array(speak_list).reshape(-1,1))
x=np.arange(np.min(stroke_list),np.max(stroke_list)+0.1,(np.max(stroke_list)-np.min(stroke_list))/10)
prediction=model.predict(x.reshape(-1,1))

plt.scatter(stroke_list,speak_list)
plt.plot(x,prediction)

plt.grid()
plt.xlabel('stroke_speed[cm/s]')
plt.ylabel('speak_speed[mora/s]')
plt.title('stroke_speak_relationship')



fig.savefig('1221.png')
```
